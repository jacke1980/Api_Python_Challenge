```python
Weather_path="../output_data/newcity.csv"
Weather_Vacation=pd.read_csv(Weather_path)
```

```python
#CONVERT UNIX TIME TO DATETIME AND ADD TO NEW COLUMN
time_date=Weather_Vacation.iloc[:,9].to_list()
times=[]
for t in time_date:
    time=datetime.utcfromtimestamp(t).strftime("%y-%m-%d")
    times.append(time)
    
Weather_Vacation.loc[:,"Time Date"]=times
```

```python
Weather_Vac=Weather_Vacation.rename(columns={"lat":"Lat", "lon":"Lon"})
Weather_Vac=Weather_Vac[["City","Country","Lat","Lon","Max Temp","Humidity","Cloudiness","Wind Speed","Date","Time Date"]]
```

```python
#HUMIDITY HEATMAP
Weather_Vac.dtypes

```

```python
locations = Weather_Vac[["Lat", "Lon"]]

locations
```

```python
Humidity = Weather_Vac["Humidity"].astype(float)
```

```python
# Plot Heatmap
fig = gmaps.figure()

# Create heat layer
heat_layer = gmaps.heatmap_layer(locations, weights=Humidity, 
                                 dissipating=False, max_intensity=50,
                                 point_radius=1)


# Add layer
fig.add_layer(heat_layer)

# Display figure
fig
```

```python
#Create new DataFrame fitting weather criteria
```

```python
hotel_df=Weather_Vac.loc[(Weather_Vac["Wind Speed"]<10)&(Weather_Vac["Cloudiness"]<1)& 
                         (Weather_Vac["Max Temp"]>70)&(Weather_Vac["Max Temp"]<80)].reset_index(drop=True)
hotel_df
```

```python
hotel_names = []
geocode_url = 'https://maps.googleapis.com/maps/api/geocode/json?'
for index, row in hotel_df.iterrows():
    params = {
        'address': row['City'],
        'key': g_key
    }
    response = requests.get(geocode_url, params).json()
    lat = response['results'][0]['geometry']['location']['lat']
    lng = response['results'][0]['geometry']['location']['lng']
    location = f'{lat},{lng}'
    keyword = '&keyword=hotel'
    nearby_url = 'https://maps.googleapis.com/maps/api/place/nearbysearch/json?'
    final_url = f'{nearby_url}&location={location}{keyword}&rankby=distance&key={g_key}'
    hotel = requests.get(final_url).json()
#     print(hotel['results'][0]['name'])
    try:
        hotel_names.append(hotel['results'][0]['name'])
    except:
        hotel_names.append('N/A')
hotel_df['Hotel Name'] = hotel_names
hotel_df
```

```python
hotel_df.to_csv("../output_data/Hotels.csv", index=True, header=True)
```
